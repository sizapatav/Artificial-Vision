## Actividad 4.

Obtenga de internet un mapa de alturas cualquiera, en escala de grise. Aplique pseudocolor basado en planos de intensidad, con por lo menos 20 niveles de color NO aleatorios.
<br>Ejemplo de mapa utilizable: https://www.leonardobenini.it/wp-content/uploads/2015/05/02-australia-gray.jpg</br>

```python
#Librerias necesarias
%matplotlib inline 
import cv2
import numpy as np
import random as rand

import math 
import matplotlib.pyplot as plt
#Se definen las funciones correspondencia y pseudocolor
def correspondence(val, paleta, rango):
    '''
    Receives the value of the grayscale, the color palette and size of the range array.
    Returns a tuple from the palette depending on the part of the range to which val belongs.
    '''
    #"Recibe el valor de la escala de grises, la paleta de colores y el tamaño del vector,
    #Devuelve una tupla de la paleta dependiendo del intervalo al cual la variable -val- pertenezca"
    
    #se realiza una normalización a todos los valores de la matriz val.
    pos=int( (val*(rango)/255))-1
    r,g,b=paleta[pos]
    return (r,g,b)

def pseudocolor(img):
    #Espacio por color
    espacio_color=7
    rango=3* espacio_color+1
    
    #Generación de matriz auxiliar
    aux=np.zeros((rango,3), dtype=np.uint8)
    vec_aux=np.zeros((espacio_color), dtype=np.uint8)
    #for i in range(0,rango):
    #    aux[i][0]=1
        
    #print(aux[1][:])
    var=2* espacio_color
    
    for d in range(espacio_color):
        vec_aux[d]=d*255/(espacio_color-1)
    #print(vec_aux)
    
    for i in range(rango):
        if i<espacio_color:
            
            aux[i][0]=255
            aux[i][2]=0
            aux[i][1]=vec_aux[i]
              
        elif i> var:
            
            aux[i][0]=0
            aux[i][2]=255
            aux[i][1]=vec_aux[rango-1-i]
            
        elif i>=espacio_color and i<=(espacio_color*3/2):
            aux[i][0]=vec_aux[espacio_color*2-i-2]
            aux[i][1]=255
            aux[i][2]=0
        elif i<=var and i>(espacio_color*3/2):
            aux[i][0]=0
            aux[i][1]=255
            aux[i][2]=vec_aux[i-espacio_color-2]
    #print(aux)
    
    vect=[tuple(aux[i][:]) for i in range(rango)]
    paleta=vect
    #print(vect)
    
    #paleta= [tuple(rand.randint(0,255) for j in range(3)) for i in range(rango)] 
    
    #generación aleatoria de la paleta de colores.
    tam=np.shape(img)
    res=np.zeros((tam[0],tam[1], 3), dtype=np.uint8)
    #se aplica a la matriz res, la función de correspondencia con la imagen original en el rango indicado
    for r in range(tam[0]):
        for c in range(tam[1]):
            res[r][c]=correspondence(img[r][c], paleta, rango)
    return res    #se devuelve la imagen res, una vez se le aplica la función correspondencia

print('Método de pseudocolor cargado')
```

```python
#imprimir la imagen en pseudocolor

img_USA =  cv2.imread("res/USA.jpg", cv2.IMREAD_GRAYSCALE)
pseudocolor_img_USA = pseudocolor(img_USA)

fig, ((ax1), (ax2)) = plt.subplots(1,2, figsize=(18, 4))

fig.suptitle("Gray Scale VS Pseudocolor", fontsize=20)
ax1.set_title('Gray Scale')
ax1.imshow(img_USA,cmap='gray', aspect='auto')
ax2.set_title('Pseudocolor')
ax2.imshow(pseudocolor_img_USA, aspect='auto')
```

```python
#imprimir la imagen en pseudocolor

img_w =  cv2.imread("res/World.png", cv2.IMREAD_GRAYSCALE)
pseudocolor_img_w = pseudocolor(img_w)

fig, ((ax1), (ax2)) = plt.subplots(1,2, figsize=(18, 4))

fig.suptitle("Gray Scale VS Pseudocolor", fontsize=20)
ax1.set_title('Gray Scale')
ax1.imshow(img_w,cmap='gray', aspect='auto')
ax2.set_title('Pseudocolor')
ax2.imshow(pseudocolor_img_w, aspect='auto')
```
